Q3. Show sentiment distribution (positive, negative, neutral). (Piechart)

```python
bin = [ 0,2,3,5]
labels1 = ['Negative', 'Neutral', 'Positive']

df['ratingKeyword']= pd.cut(df['rating'],bins = bin,labels=labels1)
```

```python
df
```

```python
plt.pie(df['ratingKeyword'].value_counts(), labels=df['ratingKeyword'].value_counts().index)
plt.title('Sentiment distribution')
```

Q4. Detect models that have price above overall mean but rating below overall mean — mark them on a scatter (Find “Overpriced” Models (High Price, Low Rating))



```python
hplr= df.groupby('model')[['price_usd','rating']].mean()
priceMean= df['price_usd'].mean()
ratMean = df['rating'].mean()


overpriced = hplr[(hplr['price_usd'] > priceMean) & (hplr['rating'] < ratMean)]
```

```python
plt.figure(figsize=(8,5))

# Plot all models
plt.scatter(hplr['price_usd'], hplr['rating'], label="Other Models", color='blue')

# Highlight overpriced ones in red
plt.scatter(overpriced['price_usd'], overpriced['rating'],
            color='red', s=100, label="Overpriced Models")

# Label axes and title
plt.xlabel("Average Price (USD)")
plt.ylabel("Average Rating")
plt.title("Price vs Rating — Highlighting Overpriced Models")

# Add mean reference lines
plt.axvline(priceMean, color='green', linestyle='--', label='Mean Price')
plt.axhline(ratMean, color='orange', linestyle='--', label='Mean Rating')

plt.legend()
plt.show()
```
